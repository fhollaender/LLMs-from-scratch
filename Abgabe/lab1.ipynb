# Modularbeit: Large Language Models from Scratch
**Gruppe:** Phillip Graf, Konstantin Schmidt, Fabian Holländer

Ziel unserer Modularbeit ist es, die Methoden aus dem Repository [rasbt/LLMs-from-scratch](https://github.com/rasbt/LLMs-from-scratch) auf eine eigene Rezept-Domäne zu übertragen. Die folgenden Datensätze bilden dafür die gemeinsame Grundlage über die späteren Labs hinweg.

### Trainings-Datensatz (RecipeNLG)

* **Name:** `{R}ecipe{NLG}: A Cooking Recipes Dataset for Semi-Structured Text Generation`
* **Beschreibung:** Ein semistrukturierter Datensatz, der speziell für das Training von Modellen zur Generierung von Kochrezepten entwickelt wurde.
* **Autoren:** Bien et al.
* **Webseite:** [RecipeNLG Projektseite](https://recipenlg.cs.put.poznan.pl)
* **Umfang:** ca. 2 Millionen Rezepte
* **Verwendung im Projekt:** Tokenisierung, Pretraining, eigener Instruction-Datensatz und RAG.
* **Download Links:**
    * [Reduzierter Datensatz mit 100k Rezepten](https://syncandshare.lrz.de/dl/fiHE8nDPcb4nww3VCn4QmN/reduced_dataset_100k.csv)
    * [Vollständiger Datensatz](https://syncandshare.lrz.de/dl/finR5gtyQx3FNL5P2hz6H/full_dataset.csv)
* **Struktur (Spalten):**
    * `title`: Der Name/Titel des Rezepts.
    * `ingredients`: Die Zutaten des Rezepts inklusive Maßeinheiten.
    * `directions`: Die Schritt-für-Schritt-Zubereitungsanleitung.
    * `link`: Die Original-URL zum Rezept.
    * `source`: Die Art der Quelle (z. B. "Gathered").
    * `NER` (*Named Entity Recognition*): Extrahierte Lebensmittelkomponenten direkt aus den Zutaten.

#### Warum haben wir uns für diesen Datensatz entschieden?

Unser ursprüngliches Ziel war es, ein Large Language Model (LLM) zur vollständigen Generierung von Kochrezepten zu entwickeln. Im Laufe des Semesters mussten wir jedoch feststellen, dass das Training eines komplett eigenen, generativen Modells von Grund auf den zeitlichen und technischen Rahmen dieses Kurses sprengen würde. Dennoch haben wir uns entschieden, weiterhin mit dem RecipeNLG-Datensatz zu arbeiten, da er eine sehr gute Datenqualität aufweist. Der Datensatz bietet folgende **Vorteile**:

* **Viele Daten:** Mit über 2 Millionen Rezepten und einer klaren, semistrukturierten Architektur bietet er mehr als genug Datenmaterial, um komplexe sprachliche Muster zu erlernen.
* **Gute Ausgangsstruktur:** Die Daten liegen bereits semistrukturiert vor. Dadurch konnten wir uns auf domänenspezifische Formatierung und Modelltraining konzentrieren.
* **Praxisnaher Aufbau:** Die klare Hierarchie (Titel, Zutaten, Anleitung) spiegelt die reale Struktur von Rezepten gut wider.

### Der Klassifikations-Datensatz (3A2M)

Diesen Datensatz verwenden wir in Kapitel 6 für das Fine-Tuning eines bereits vortrainierten OpenAI-GPT-2-Modells, um Rezepte nach den nachfolgenden Kategorien zu klassifizieren.

* **Name:** `3A2M Cooking Recipe Dataset`
* **Hintergrund:** "Assorted, Archetypal, and Annotated Two Million". Drei menschliche Experten haben 300.000 Rezepte aus RecipeNLG von Hand klassifiziert. Die restlichen 1,9 Millionen Rezepte wurden mittels *Active Learning* (Query-by-Committee) automatisiert gelabelt.
* **Download-Quelle:** [Kaggle - 3A2M Cooking Recipe Dataset](https://www.kaggle.com/datasets/nazmussakibrupol/3a2m-cooking-recipe-dataset)
* **Umfang:** ca. 2 Millionen Rezepte
* ****Download Links:**
    * [Reduzierter Klassifikationsdatensatz mit 100k Rezepten](https://syncandshare.lrz.de/dl/fiWvy4z3fJ2sRSLMTdQiay/reduced_classification_dataset_100k.csv)
    * [Vollständiger Klassifikationsdatensatz](https://syncandshare.lrz.de/dl/fiGzZTKfHqLLLbK2e6Uop9/full_classification_dataset.csv)

#### Die originalen Spalten (Features):
* `title`: Der Name des Gerichts (direkt aus RecipeNLG).
* `directions`: Die Schritt-für-Schritt-Zubereitungsanleitung (direkt aus RecipeNLG).
* `NER`: Die aus den Zutaten extrahierten Kern-Lebensmittelkomponenten (Named Entities).
* `genre`: Die zugewiesene Kategorie als Text-String (z. B. `bakery`, `drinks`).
* `label`: Die numerische Repräsentation des Genres (Zahlenwert für das Modell).

#### Die 9 offiziellen Genres im 3A2M-Datensatz:

Die Experten haben den Datensatz exakt in diese 9 Klassen unterteilt:

1. **`bakery`** (Backwaren, oft gebacken oder frittiert)
2. **`drinks`** (Getränke und flüssige Zubereitungen)
3. **`non-veg`** (Fleisch-, Geflügel- und Fischgerichte)
4. **`vegetables`** (Reine vegetarische Gerichte)
5. **`fast food`** (Schnellgerichte, nicht für lange Lagerung gedacht)
6. **`cereals`** (Gerichte aus Mais, Weizen, Reis oder direkt aus Getreidekörnern)
7. **`meals`** (Vollwertige Hauptmahlzeiten)
8. **`sides`** (Beilagen, Soßen und Toppings)
9. **`fusion`** (Mischformen, bei denen sich die Experten nicht einig waren)

### Der Instruction-Finetuning-Datensatz

#### Für Kapitel 7

* **Name:** `LIMA (Less Is More for Alignment)`
* **Hintergrund:** Dieser Datensatz hat keine spezifische Domain und beinhaltet Texte aus verschiedenen Themengebieten. Der Datensatz enthält sorgfältig ausgewählte Prompts und Antworten, die aus Plattformen wie Stack Exchange, wikiHow sowie von Experten erstellten Beispielen stammen.
* **Download-Quelle:** [GAIR/lima auf Hugging Face](https://huggingface.co/datasets/GAIR/lima)
* **Verwendete Projektdateien:** `mixed_train.jsonl` und `mixed_test.jsonl` unter `datasets/instruction_finetuning/`.

#### Für Kapitel 7b

* **Name:** `EID (Eigener Instruction-Datensatz)`
* **Hintergrund:** Dieser Datensatz basiert auf RecipeNLG und wurde von uns selbst zusammengestellt. Er umfasst 17.000 Beispiele: 15.000 Trainingsbeispiele, 1.000 Validierungsbeispiele und 1.000 Testbeispiele im Phi-3-nahen Chatformat.
* **Erzeugung:** [instruction_dataset.ipynb](./instruction_dataset.ipynb)
* **Direkte Download-Links:**
    * [Train](https://syncandshare.lrz.de/dl/fi7TGWhZKqmjF8Wsn9Ze9j/recipe_instruct_train_ctx512_seed123.jsonl)
    * [Validation](https://syncandshare.lrz.de/dl/fi89FcTYVkG4jNacPVJVMx/recipe_instruct_val_ctx512_seed123.jsonl)
    * [Test](https://syncandshare.lrz.de/dl/fiBKi9JVNvco2mPE1Kzgp5/recipe_instruct_test_ctx512_seed123.jsonl)

Beispiel:
> <|user|>  
> "All I've got is shrimp, mayo, onion, cream cheese, lemon juice, celery. What can I cook?"
>
> <|assistant|>

> Ingredients:  
> 8 oz. cream cheese, softened  
> 1/2 c. mayo  
> 1/4 c. chopped onion  
> 1 rib chopped celery  
> 1 tsp. lemon juice

### RAG-Datensatz

Für Kapitel 8 verwenden wir keinen neuen Rohdatensatz, sondern denselben RecipeNLG-basierten Rezeptkorpus wie in den vorherigen Labs. Der Unterschied liegt in der Aufbereitung: Für Retrieval-Augmented Generation behandeln wir jedes Rezept als eigenen Retrieval-Chunk mit Titel, Zutaten und Zubereitungsschritten.

* **Datengrundlage:** RecipeNLG
* **Demo-Umfang:** die ersten 1.000 Rezept-Chunks, damit Embedding-Erzeugung und lokale Ausführung reproduzierbar bleiben.
* **Embedding-Modell:** [hf.co/CompendiumLabs/bge-base-en-v1.5-gguf](https://huggingface.co/CompendiumLabs/bge-base-en-v1.5-gguf)
* **Sprachmodell:** [hf.co/bartowski/Llama-3.2-1B-Instruct-GGUF](https://huggingface.co/bartowski/Llama-3.2-1B-Instruct-GGUF)
* **Lokales Tooling:** [Ollama](https://ollama.com/)

Die RAG-Pipeline und die konkrete Verarbeitung werden in [lab8.ipynb](./lab8.ipynb) beschrieben.
