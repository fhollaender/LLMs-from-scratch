# University Project: Large Language Models from Scratch
**Group:** Phillip Graf, Konstantin Schmidt, Fabian Holländer

**Dataset:** {R}ecipe{NLG}: A Cooking Recipes Dataset for Semi-Structured Text Generation
- A semi-structured dataset for training models to generate cooking reciepes
- Authors: Bien et al. 
- https://recipenlg.cs.put.poznan.pl
- ~ 2 Million Reciepes
- Semi-Structured with the columns: 
    - *title*: Title of the reciepe
    - *ingredients*: Ingriedients of the reciepe with unit of measurement
    - *directions*: Instruction of the reciepe.
    - *link*: URL to the reciepe
    - *source*: Type of source e.g. "Gathered"
    - *NER*: Named Entity Recognition, food components extracted from the ingredients.

**Why did we choose this dataset?** 

Our goal was to develop an LLM for generating cooking recipes. During the course, we came to the conclusion that training a full generative model from scratch was beyond the scope of the course. However, we decided to proceed with the RecipeNLG dataset due to its exceptional quality. The dataset offers the following **advantages**:
- **Large size**: With over 2.2 million recipes and a semi-structured architecture, the dataset provides ample data for learning complex language patterns.
- The dataset is already **cleaned and preprocessed**, saving significant time and errors while training the model. Only minor additional preprocessing steps were required on our part (see following chapters).
- The clear hierarchy (Title, Ingredients, Directions) accurately reflects the **real-world structure of recipes**.
