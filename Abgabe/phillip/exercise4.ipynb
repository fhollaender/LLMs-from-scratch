Interessante Beobachtung:
- erkennt die wörter als einzelne tokens
- random wort "kjs" wird aufgeteilt in 2 tokens

```python
import tiktoken

tokenizer = tiktoken.get_encoding("gpt2")

batch = []

txt1 = "Every effort moves us sunset kjs"
txt2 = "Every day holds a sun rise ha"

batch.append(torch.tensor(tokenizer.encode(txt1)))
batch.append(torch.tensor(tokenizer.encode(txt2)))
batch = torch.stack(batch, dim=0)
print(batch)
```
